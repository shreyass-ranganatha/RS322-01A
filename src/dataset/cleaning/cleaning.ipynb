```python
from collections import defaultdict
import cv2
import json
import os
import glob
```

```python
imgsdir = "/Users/shreyas/Developer/Research/GroundingDINO/data/satellite/images"
boxsdir = "/Users/shreyas/Developer/Research/GroundingDINO/data/satellite/boxes"
```

```python
boxs_yolo = {}
boxs_watershed = {}

for f in glob.glob(os.path.join(boxsdir, "*.json")):
    name = os.path.basename(f)
    dt = json.load(open(f))

    if not isinstance(dt, dict):
        continue

    match dt["type"]:
        case "yolo":
            boxs_yolo[name.replace(".json", "")] = dt["box"]

        case "watershed":
            boxs_watershed[name.replace(".json", "")] = dt["box"]
```
